```python
import numpy as np
import numpy.linalg as la

np.set_printoptions(precision=3, linewidth=100, threshold=30)
n = 200
```

```python
scale = 10**np.linspace(0, -12, n)
D = -2*np.diag(scale) + np.diag(scale[:-1], 1) + np.diag(scale[1:], -1)
D
```

```python
la.cond(D, 2)
```

```python
xtrue = np.random.randn(n)
b = D @ xtrue
b
```

```python
xhat = la.solve(D, b)
```

Now fix the row scaling in $A$:

```python
#clear
M = np.diag(1/scale)
print(M@D)
```

And solve the modified linear system:

```python
#clear
x = la.solve(M@D, M@b)
```

```python
la.norm(xtrue-x, 2)/la.norm(x, 2), la.norm(xtrue-xhat, 2)/la.norm(x, 2)
```

What do you think the residuals will be in each case?

```python
la.norm(D@x - b, 2)/la.norm(b, 2), la.norm(D@xhat - b, 2)/la.norm(b, 2)
```
