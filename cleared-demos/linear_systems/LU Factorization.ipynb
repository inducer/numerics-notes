Recall the "recipe" for LU factorization:

$$\let\B=\boldsymbol \begin{array}{cc}
     & \left[\begin{array}{cc}

     \end{array}\right]\\
     \left[\begin{array}{cc}
       1 & \\
       \B{l}_{10} & L_{11}
     \end{array}\right] & \left[\begin{array}{cc}
       a_{00} & \B{a}_{01}\\
       \B{a}_{10} & A_{11}
     \end{array}\right]
   \end{array}$$

Find $u_{00}$ and $u_{01}$. Check `A - L@U`.
